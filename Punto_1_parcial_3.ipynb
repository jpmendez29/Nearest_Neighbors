# punto 1 - KNN

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.datasets import load_digits
from sklearn import neighbors
from sklearn.metrics import accuracy_score, confusion_matrix, recall_score, precision_score
```

# Data set 

image.png

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_digits.html#sklearn.datasets.load_digits

```python
#obtenemos los datos 
digits = load_digits()
```

```python
np.random.seed(100)
```

```python
X = digits.data
y = digits.target
```

```python
# divicion datos 70%-30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1970)
```

```python
# Knn con numero de vecinos de 100 con una metrica de minkowski

clfKNN = neighbors.KNeighborsClassifier(n_neighbors=50, # número de vecinos
                                        weights='distance', # pesos asignados a los vecinos al momento de tomar la decisión
                                                           # 'uniform', 'distance'
                                        algorithm='brute', # optimiza la ejecución del clasificador
                                                           # ‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’
                                        p=2, # usado para la fórmula de la distancia de Minkowski
                                        metric='minkowski')
clfKNN.fit(X_train,y_train)
y_esperado = y_test
y_predicho = clfKNN.predict(X_test)
```

```python
# Medidas de desempeño 

print(accuracy_score(y_esperado,y_predicho))
print(recall_score(y_esperado,y_predicho, average = 'weighted'))
print(precision_score(y_esperado,y_predicho, average = 'weighted'))
print(confusion_matrix(y_esperado,y_predicho))
```

Podemos ver que el sistema clasifica correctamente un 95% de los datos de prueba 

# RNN

```python
# divicion datos 70%-30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1970)
```

```python
# Rnn con un radio de 0.5 con una metrica de minkowski

clfRNN = neighbors.RadiusNeighborsClassifier(radius=0.5, # número de vecinos
                                        weights='uniform', # pesos asignados a los vecinos al momento 
                                                           # de tomar la decisión
                                                           # 'uniform', 'distance'
                                        algorithm='brute', # optimiza la ejecución del clasificador
                                                           # ‘auto’, ‘ball_tree’, ‘kd_tree’, ‘brute’
                                        p=2, # usado para la fórmula de la distancia de Minkowski
                                        metric='minkowski',
                                        outlier_label=0
                                        #outlier_label='most_frequent'
                                        )
clfRNN.fit(X_train,y_train)
y_esperado = y_test
y_predicho = clfRNN.predict(X_test)
```

```python
# Medidas de desempeño 

print(accuracy_score(y_esperado,y_predicho))
print(recall_score(y_esperado,y_predicho, average = 'weighted'))
print(precision_score(y_esperado,y_predicho, average = 'weighted'))
print(confusion_matrix(y_esperado,y_predicho))
```

Podemos ver que el sistema clasifica correctamente un 10% de los datos de prueba
